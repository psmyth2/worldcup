```python
import json
import pandas as pd
```

```python
def pop_tweets(path):
    #Declare a new data frame with pandas, with some specific column names
    tweets = pd.DataFrame(columns=['screenName','userId','text','location'])

    #Open the text file that contains the tweets we collected
    tweets_file = open(path, "r")
    
    #Read the text file line by line
    for line in tweets_file:
        
        #Load the JSON information
        tweet = json.loads(line)
        
        #If the tweet isn't empty, add it to the data frame
        if ('text' in tweet): 
            tweets.loc[len(tweets)]=[tweet['user']['screen_name'],\
            tweet['user']['id'],tweet['text'],tweet['place']['full_name']]    

    return tweets

WorldCup_tweets = pop_tweets('worldcupsentiments.json')
```

```python
#Remove tweets that discuss two or more teams
for row in range(len(WorldCup_tweets)):
    
    #Set three flags that determine if the tweet mentions each team
    has_f=any(word in WorldCup_tweets.loc[row,'text'].lower() for word in \
    	['france','#france','#fra'])
    has_c=any(word in WorldCup_tweets.loc[row,'text'].lower() for word in \
    	['croatia','#croatia','#cro'])
    has_e=any(word in WorldCup_tweets.loc[row,'text'].lower() for word in \
    	['england','#england','#eng'])
    has_b=any(word in WorldCup_tweets.loc[row,'text'].lower() for word in \
    	['belgium','#belgium','#bel'])
    
    #If the tweet mentions more than one team, set the new 'multi-team' column to True
    if has_f+has_c+has_e+has_b > 1:
        WorldCup_tweets.loc[row,'multi-team']=True
    else:
        WorldCup_tweets.loc[row,'multi-team']=False 
```

```python
#Applying cut to remove multi-team tweets
WorldCup_tweets = WorldCup_tweets[WorldCup_tweets['multi-team'] == False]
WorldCup_tweets.head(n=5)
```

```python
WorldCup_tweets = WorldCup_tweets.reset_index(drop=True)
WorldCup_tweets.head(n=20)
```
